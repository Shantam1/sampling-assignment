```python
# simple random sampling
n= int((1.96*1.96 * 0.5*0.5)/(0.05**2))
Sampled_df = New_df.sample(n=n, random_state=42)
print(Sampled_df)
```

```python
# systematic sampling

Interval = 2
Systematic_df = New_df.iloc[::Interval]
Systematic_df
```

```python
# stratified sampling

from sklearn.model_selection import train_test_split

n = int((1.96*1.96 * 0.5*0.5)/((0.05)**2))

Strata = New_df.groupby('Class')

# sample 2 rows from each stratum
Stratified_df = Strata.apply(lambda x: x.sample(n))
Stratified_df
```

```python
# cluster sampling
import numpy as np

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10, random_state=42).fit(New_df)
cluster_assignments = kmeans.labels_

# Select the clusters you want to include in the sample
selected_clusters = [0, 2, 4, 5, 8]

cluster_series = pd.Series(cluster_assignments)

# Create the new DataFrame containing only the rows from the selected clusters
df_cluster_sample = New_df[cluster_series.isin(selected_clusters)]


# Print the resulting DataFrame
(df_cluster_sample)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Split the data into training and testing sets
train_dfs = []
test_dfs = []
combinedsampledf=[Systematic_df,Sampled_df,Stratified_df,Convenience_sample,df_cluster_sample]
for dff in combinedsampledf:
    X = dff.drop('Class', axis=1)
    y = dff['Class']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    train_dfs.append(pd.concat([X_train, y_train], axis=1))
    test_dfs.append(pd.concat([X_test, y_test], axis=1))

# Initialize the models
dtc = DecisionTreeClassifier()
knn = KNeighborsClassifier()
gnb = GaussianNB()
rfc = RandomForestClassifier()
svm = SVC()
# Train and test each model
results = []
modname=["Decision Tree Classifier","KNN Classifier","Gaussian Naive Bayes","Random Forrest Classifier","Support Vector Machine"]
sname=["Simple Random Sampling","Systematic Sampling","Stratified Sampling","Cluster Sampling","Convenience Sampling"]
for i in range(len(combinedsampledf)):
    train_df = train_dfs[i]
    test_df = test_dfs[i]
    model_results = []
    x=0
    for model in [dtc, knn, gnb, rfc, svm]:
        #model_name = type(model)._name_
        model.fit(train_df.drop('Class', axis=1), train_df['Class'])
        accuracy = model.score(test_df.drop('Class', axis=1), test_df['Class'])
        model_results.append([modname[x],accuracy])
        x+=1
    results.append(pd.DataFrame(model_results, columns=['Model','Accuracy']))

# Concatenate the results into a single DataFrame
all_results = pd.concat(results, keys=[i for i in sname])
```
